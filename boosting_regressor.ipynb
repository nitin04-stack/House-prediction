```python
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.datasets import make_regression
from sklearn.metrics import r2_score,accuracy_score
```

```python
x,y = make_regression(
    n_samples = 1000,
    n_features = 10,
    noise = 20,
    random_state = 42
)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr = GradientBoostingRegressor(
    n_estimators = 200,
    learning_rate = 0.05,
    max_depth = 3,
    subsample = 0.8,
    random_state = 42
)
```

```python
gbr.fit(x_train,y_train)
```

```python
y_pred = gbr.predict(x_test)
```

```python
r2_score(y_pred,y_test)
```

# classification

```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(
    n_samples = 500,
    n_features = 20,
    n_informative = 10,
    random_state = 42
)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 42)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier(
    n_estimators = 201,
    learning_rate = 0.1,
    max_depth = 3,
    random_state = 42
)
```

```python
gbc.fit(x_train,y_train)
```

```python
y_pred = gbc.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np

toxidromes = ['Cholinergic', 'Sedative', 'Opioid', 'Anticholinergic', 
              'Sympathomimetic', 'Mixed', 'Serotonin', 'Hallucinogenic']
metrics = ['Sensitivity', 'Specificity', 'F1-Score', 'AUROC']
data = np.array([
    [0.971, 0.992, 0.968, 0.997],
    [0.944, 0.978, 0.943, 0.988],
    [0.938, 0.981, 0.941, 0.986],
    [0.921, 0.974, 0.928, 0.981],
    [0.912, 0.969, 0.919, 0.977],
    [0.882, 0.961, 0.881, 0.963],
    [0.903, 0.975, 0.906, 0.979],
    [0.858, 0.952, 0.867, 0.944]
])

x = np.arange(len(toxidromes))
width = 0.2

fig, ax = plt.subplots(figsize=(14, 8))
ax.bar(x - 1.5*width, data[:,0], width, label='Sensitivity', alpha=0.8)
ax.bar(x - 0.5*width, data[:,1], width, label='Specificity', alpha=0.8)
ax.bar(x + 0.5*width, data[:,2], width, label='F1-Score', alpha=0.8)
ax.bar(x + 1.5*width, data[:,3], width, label='AUROC', alpha=0.8)

ax.set_xlabel('Toxidrome Classes', fontsize=12)
ax.set_ylabel('Performance Score', fontsize=12)
ax.set_title('Toxidrome Classification Model Performance\n(Grouped Bar Histogram)', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(toxidromes, rotation=45, ha='right')
ax.legend()
ax.set_ylim(0.8, 1.0)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
